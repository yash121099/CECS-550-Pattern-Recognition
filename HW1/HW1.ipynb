#Part 3.1

**Identity Matrix:**

An identity matrix is a square matrix in which all the elements of the principal diagonal that is from top-left to bottom-right are ones, and all other elements are zeros.

**Effect of Multiplying with an Identity Matrix:**

Multiplying any matrix A with an identity matrix of appropriate dimensions results in the matrix A itself. In mathematical terms, for any matrix A:

A * I = I * A = A

The identity matrix is the neutral element in matrix multiplication. Any matrix multiplied by the identity matrix remains unchanged.

**Citation:**

Lay, D. C., Lay, S. R., & McDonald, J. J. (2016). Linear algebra and its applications (5th ed.). Pearson.

## Extra Points

* Now can you perform the scrambling on __both the columns and rows__ of the original image?  Write the code and plot the scrambled image. Then invert the operations and show that you can recover the original image.

```python
# Recreate the identity matrix and shuffled rows
identity_matrix = np.eye(image.shape[0])
indices = np.arange(image.shape[0])
np.random.shuffle(indices)
rows = indices

# reCreate matrix using shuffled rows
T = identity_matrix[rows]

# Compute the inverse of the matrix T
Tinv = np.linalg.inv(T)

# Double scramble both the columns and rows of the original colored image
double_scrambled_colored_image = np.zeros_like(image)
for channel in range(3):
    double_scrambled_colored_image[:, :, channel] = T @ image[:, :, channel] @ T.T

# showing the scrambled colored image(both rows and coloumn that is double scrambled image) that is of the orignal image
plt.imshow(double_scrambled_colored_image.astype(np.uint8))
plt.title('Row and Column-wise double Scrambled Image')
plt.axis('off')
plt.show()
```

```python
# Apply the inverse transformations to both rows and columns to retrieve the original image
retrieved_double_scrambled_image = np.zeros_like(image)
for channel in range(3):
    retrieved_double_scrambled_image[:, :, channel] = Tinv @ double_scrambled_colored_image[:, :, channel] @ Tinv.T

# Plot the retrieved colored image after double scrambling
plt.imshow(retrieved_double_scrambled_image.astype(np.uint8))
plt.title('Retrieved Image after Double Scrambling')
plt.axis('off')
plt.show()
```
